```python
import pandas as pd

dt_pr = pd.read_csv('result.csv')
dt_pr
```

```python
df = dt_pr.drop(columns=['annotator', 'annotation_id', 'created_at', 'lead_time', 'updated_at'], axis=1)

df.head(5)
```

```python
# 1. Проверяем столбцы
print("Столбцы:", df.columns.tolist())

# 2. Пусть исходный столбец называется 'choices'
source_col = "categories"  # измените на "categories", если так у вас

# 3. Список всех реакций
columns = [
    "Нравится скорость отработки заявок",
    "Нравится качество выполнения заявки",
    "Нравится качество работы сотрудников",
    "Понравилось выполнение заявки",
    "Вопрос решен"
]

# 4. Приводим к строкам и убираем NaN
df[source_col] = df[source_col].fillna("").astype(str)

# 5. Для каждой реакции создаём колонку с 1/0 по вхождению подстроки
for col in columns:
    df[col] = df[source_col].str.contains(col, regex=False).astype(int)

# 6. Результат
print(df.head())
```

```python
df = df.to_csv('Test данные.csv', index=False)
```
